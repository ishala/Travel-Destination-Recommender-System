```python
#import sklearn library
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

data = pd.read_excel('/content/datawisata.xlsx')
# Use pd.read_excel to read Excel files

# prompt: make code for item collaborative filtering using rating and types features from datawisata

import pandas as pd
# Create a user-item matrix
# The 'business_id' column is likely to be unique for each business, so we'll use it as the index.
user_item_matrix = data.pivot_table(index='types', columns='name', values='rating')
```

```python
user_item_matrix
```

```python

# Fill NaN values with 0
user_item_matrix = user_item_matrix.fillna(0)
user_item_matrix
```

```python


# Calculate cosine similarity between items
item_similarity = cosine_similarity(user_item_matrix.T)

# Convert similarity matrix to DataFrame
item_similarity_df = pd.DataFrame(item_similarity, index=user_item_matrix.columns, columns=user_item_matrix.columns)

# Function to get item recommendations
def get_item_recommendations(item_name, top_n=3):
  """
  Get item recommendations based on item collaborative filtering.

  Args:
    item_name: Name of the item for which to find recommendations.
    top_n: Number of recommendations to return.

  Returns:
    A list of top_n recommended items.
  """
  similar_items = item_similarity_df[item_name].sort_values(ascending=False)[1:top_n+1]
  return similar_items.index.tolist()

# Example usage
recommendations = get_item_recommendations('Pantai Parangkusumo', top_n=10)
print(f"Recommendations for : {recommendations}")
```
